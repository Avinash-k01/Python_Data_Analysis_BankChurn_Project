Data Visualization:

```python
# Churn rate calculation
churn_rate = bank['Attrition_Flag'].value_counts(normalize=True) * 100
sns.barplot(churn_rate)
plt.title('Churn_rate based on Attrition_Flag')
```

```python
#distribution of churn across different demographic factors
# Age distribution for churners vs non-churners
sns.boxplot(x='Attrition_Flag', y='Customer_Age', data=bank)
plt.title('Customer Age Distribution by Churn Status')
plt.show()
```

```python
# Churn rate by gender
churn_by_gender = bank.groupby('Gender')['Attrition_Flag'].value_counts(normalize=True).unstack() * 100
# Reset the index and melt the dataframe for easier plotting
churn_by_gender = churn_by_gender.reset_index().melt(id_vars='Gender', var_name='Attrition', value_name='Percentage')

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Percentage', hue='Attrition', data=churn_by_gender)

# Add title and labels
plt.title('Churn Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.show()
```

```python
# Compare transaction amounts for churners vs non-churners
sns.boxplot(x='Attrition_Flag', y='Total_Trans_Amt', data=bank)
plt.title('Total Transaction Amount by Churn Status')
plt.show()

# Analyze months inactive for churners vs non-churners
sns.boxplot(x='Attrition_Flag', y='Months_Inactive_12_mon', data=bank)
plt.title('Months Inactive by Churn Status')
plt.show()
```

```python
# Analyze credit utilization for churners vs non-churners
sns.barplot(x='Attrition_Flag', y='Avg_Utilization_Ratio', data=bank)
plt.title('Credit Utilization Ratio by Churn Status')
plt.show()

# Credit limit comparison
sns.barplot(x='Attrition_Flag', y='Credit_Limit', data=bank)
plt.title('Credit Limit by Churn Status')
plt.show()
```

```python
#correlation between numeric data type columns
# Correlation matrix
correlation_matrix = bank[['Customer_Age','Dependent_count','Months_on_book','Total_Relationship_Count','Months_Inactive_12_mon']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Distribution of Education Level
bank.groupby('Education_Level').size().sort_values(ascending=False).plot(kind='bar')
plt.title('Distribution of Education Level')
plt.xticks(rotation=45)
```

```python
#Distribution of Marital status
bank.groupby('Marital_Status').size().sort_values(ascending=False).plot(kind='bar')
plt.title('Distribution of Marital_Status')
plt.xticks(rotation=45)
```

Insight Generation:

-->Characteristics of Customers Likely to Churn:

Based on the analysis of your bank churners dataset, the following key characteristics have emerged as strong indicators of customer churn:

1. Demographic Patterns:

    Age: Customers on the younger (e.g., below 30) and older (e.g., above 60) ends of the age spectrum are more likely to churn. Middle-aged customers (30-50) tend to remain more loyal.

    Income Category: Lower-income customers tend to have higher churn rates, while higher-income customers show more loyalty. This pattern suggests that financial stability plays a role in retention.
    
    Education Level: Customers with lower educational attainment may exhibit higher churn rates. In contrast, more educated customers, especially those with college degrees or higher, may have more stable relationships with the bank.

2. Behavioral and Engagement Factors:

    Transaction Activity:
        Low Transaction Volume: Customers with lower total transaction amounts or lower transaction counts are more likely to churn. Engaged customers who use the bank's services frequently tend to remain.

    Months of Inactivity: Customers who have had multiple months of inactivity  are at higher risk of churn.
   
3. Financial Behavior:

    Credit Utilization: Customers with higher credit utilization ratios (e.g., utilizing over 80% of their available credit) are more likely to churn. This suggests financial stress, as customers maxing out their credit lines could be under pressure, making them more prone to leaving.
    
    Total Revolving Balance: Higher revolving balances (i.e., the amount of unpaid balance that rolls over) can be an indicator of customers who are at risk. They may struggle with repayments and therefore churn if they feel overburdened.

    Credit Limit: Customers with lower credit limits seem more likely to churn. Offering higher credit limits to loyal customers could improve satisfaction and retention.

4. Duration with the Bank:

Months on Book: Customers with shorter tenures at the bank (e.g., less than 1-2 years) are more likely to churn, as they may not have formed strong relationships with the bank yet. Long-term customers tend to show loyalty.

-->Patterns Based on Customer Demographics:

1)Age-Related Trends:

    Younger Customers: Typically less financially stable, may have fewer banking needs or may switch frequently between banks.
    Older Customers: While financially more stable, they may churn due to dissatisfaction with digital services or being approached by competitors with better retirement-related services.

2)Gender-Based Patterns:

    Gender differences in churn might be minor, but certain trends could emerge based on lifestyle factors. For instance, male customers might be more inclined to switch due to competitive financial offers, while female customers might prioritize consistent customer service and loyalty programs.

3)Income and Education:

    Lower-income, less-educated customers: May have higher churn rates due to financial stress, feeling underserved, or lack of engagement by the bank.

    Higher-income, well-educated customers: Likely to stay loyal, especially if offered premium banking services, higher credit limits, and personalized product offerings.


-->Key Factors Influencing Customer Churn:

1)Inactivity: Multiple months of inactivity is one of the strongest predictors of churn. Engaged customers who frequently use their credit cards or bank accounts are less likely to leave.

2)High Credit Utilization: High credit utilization is a sign of financial stress and a key indicator of customers likely to churn.

3)Transaction Volume: Lower transaction amounts and counts indicate reduced engagement and higher churn risk.

4)Low Credit Limits: Customers with low credit limits may feel dissatisfied with the bank's trust in them and switch to a competitor offering higher limits.

5)Frequent Contacts with Bank: Customers who contact the bank often (especially with complaints or inquiries) may be showing signs of dissatisfaction, increasing their likelihood of leaving.
